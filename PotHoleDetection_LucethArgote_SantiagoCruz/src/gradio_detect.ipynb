```python
import gradio as gr
from ultralytics import YOLO
import numpy as np
from PIL import Image
```

```python
model = YOLO("/content/drive/MyDrive/PotHoleDetection_LucethArgote_SantiagoCruz/src/yolov8s_finetuned1.pt")
```

```python
def yolo_inference(image):
    # Convertir PIL a numpy array RGB
    img = np.array(image)

    # Ejecutar predicción con el modelo (ajusta conf si quieres)
    results = model.predict(img, conf=0.7)

    # Usar el método plot para obtener la imagen con predicciones dibujadas
    # results.plot() devuelve un numpy array en RGB
    img_with_preds = results[0].plot()  # asumiendo batch size 1 y accediendo al primer resultado

    # Convertir a PIL Image para gradio
    return Image.fromarray(img_with_preds)

interface = gr.Interface(
    fn=yolo_inference,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="pil"),
    title="Detección de Huecos con YOLOv8",
    description="Carga una imagen y el modelo detectará huecos en la vía."
)

interface.launch(share=True)
```
